```python
# import cellmemory
from cellmemory import cm
import numpy as np
import pandas as pd
import scanpy as sc
import os
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
```

# Train model

CellMemory training includes two modes: **CellMemory** and **CellMemory-fast**  
**CellMemory**: Full -precision training that considers the expression values of all genes, including genes with zero expression.  
**CellMemory-fast**: Mixed-precision training that filters genes with zero expression in each cell.  

The following case uses the CellMemory-fast mode (use_amp=True, mode='fast') and the first fold.  
The mHypoMap data path: https://figshare.com/account/projects/241457/articles/28605038

```python
# train
batch_size = 512
Dataset = 'benchmark_mHypoMap_hvg4k_train.h5ad'  # train data path
Label = 'C66_named'   # cell type
Project = 'demo_mHypoMap_fast_valid0'  # Project name
cm.train(Dataset, Label=Label, Project=Project, batch_size=batch_size,exp_bin='bin_norm', use_amp=True, mode='fast', valid_id=0)
```

# Inference
During the inference, the parameter settings must remain consistent with those used during the training.

```python
# > inference
batch_size = 256
test_data = sc.read_h5ad('benchmark_mHypoMap_hvg4k_test.h5ad')
test_data = test_data[np.random.choice(test_data.shape[0], test_data.shape[0], replace=False), :]

adata_pred = cm.generate(test_data, Project=Project, batch_size=batch_size, out_tag=False,exp_bin='bin_norm', mode='fast')
```

```python
cls = adata_pred.cls
print('Acc:', metrics.accuracy_score(test_data.obs[Label], cls.obs['cm_pred']))
print('F1:', metrics.f1_score(test_data.obs[Label], cls.obs['cm_pred'], average="macro"))
```

# To generate UMAP visualization.

```python
# plot
sc.pp.pca(cls)
sc.pp.neighbors(cls)
sc.tl.umap(cls)

os.chdir(Project)
sc.set_figure_params(figsize=(5, 5))
```

```python
sc.pl.umap(cls, color='cm_pred', frameon=False,show=False,use_raw=False,size=5)
plt.title('CellMemory', fontsize=20)
plt.savefig('cm_cls_pred.png', dpi=300, bbox_inches="tight")
```

```python
cls.obs['Dataset'] = test_data.obs['Dataset']

sc.pl.umap(cls, color='Dataset', frameon=False,show=False,use_raw=False,size=5)
plt.title('Dataset', fontsize=20)
plt.savefig('cm_cls_dataset.png', dpi=300, bbox_inches="tight")
```
